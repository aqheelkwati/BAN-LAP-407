# bar plot

```python
sub=['maths','English','science','Social studies','Computer']
marks=[89,90,45,78,99]
```

```python
plt.figure(figsize=(8,2))
plt.bar(sub,marks)
plt.show()
```

```python
plt.figure(figsize=(8,2))
plt.bar(sub,marks,color='green')
plt.show()
```

```python
plt.figure(figsize=(8,2))
colors=['red','blue','purple','green','orange']
plt.bar(sub,marks,color=colors)
plt.show()
```

```python
plt.figure(figsize=(8,2))
plt.bar(sub,marks,color=colors,width=0.4,edgecolor='white')
plt.show()
```

```python
plt.barh(sub,marks,color=colors)
plt.show()
```

## Ploting 2 bar plots in same graph

```python
marks2=[78,56,34,90,12]

plt.bar(sub,marks)
plt.bar(sub,marks2)

plt.xlabel('subject')
plt.ylabel('Marks')
plt.show()
```

```python
sub_length=np.arange(len(sub))
wid=0.4

plt.bar(sub_length,marks,width=wid)
plt.bar(sub_length+wid,marks2,width=wid)

plt.xlabel('subject')
plt.ylabel('Marks')
plt.show()
```

```python
sub_length=np.arange(len(sub))
wid=0.4

plt.bar(sub_length,marks,width=wid,color=colors)
plt.bar(sub_length+wid,marks2,width=wid,color=colors,alpha=0.4)

plt.xlabel('subject')
plt.ylabel('Marks')
plt.show()
```

## Histogram plotting

```python
marks_50=np.random.randint(0,100,(50))
marks_50
```

```python
plt.figure(figsize=(4,4))
plt.hist(marks_50)
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(4,4))
plt.hist(marks_50,bins=bins)

plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(8,4))
plt.hist(marks_50,bins=bins,color='orange',rwidth=0.6)
plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(8,4))
plt.hist(marks_50,bins=bins,color='orange',orientation='horizontal')
plt.yticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(8,4))
plt.hist(marks_50,bins=bins,color='orange',orientation='horizontal',rwidth=0.6)
plt.yticks(np.arange(0,100,5))
plt.show()
```

```python
# Hist styles - 'bar' ,'barstacked','step','stepfilled'
bins=np.arange(0,100,5)
plt.figure(figsize=(8,4))
plt.hist(marks_50,bins=bins,color='orange',histtype='step',rwidth=0.6)
plt.xticks(np.arange(0,100,5))
plt.show()
```

### To plot 2 histographs in same plot

```python
marks_50_2=np.random.randint(0,100,(50))

bins=np.arange(0,100,5)
plt.hist([marks_50,marks_50_2],bins=bins,color=['red','blue'])

plt.xticks(np.arange(0,100,5))

plt.xlabel('marks')
plt.ylabel('Frequency')
plt.title('Marks of students of 2 classes')
plt.show()
```

# Pie Plots

```python
classes=['Physics','Chemistry','Maths','Bilogy','Statistics']
marks=[97,94,45,65,69]
```

```python
plt.pie(marks,labels=classes)
plt.show()
```

```python
colors=['red','blue','green','orange','#568474']
plt.pie(marks,labels=classes,colors=colors)
plt.show()
```

```python
plt.pie(marks,labels=classes,colors=colors,autopct='%0.3f%%')
plt.show()
```

```python
explode_values=[0.1,0,0,0,0]
plt.pie(marks,labels=classes,colors=colors,autopct='%0.3f%%',explode=explode_values)
plt.legend()
plt.show()
```

```python
plt.pie(marks,labels=classes,colors=colors,autopct='%0.3f%%',explode=explode_values, shadow=True,radius=1.4)
plt.show()
```

# Subplots

```python
study_hrs=[2,3,4,3,5,4,4]
marks=[7,17,45,34,67,32,45]
```

```python
plt.figure(figsize=(4,4))
plt.scatter(study_hrs,marks)
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.plot(study_hrs,marks,'r--')
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.hist(marks)
```

```python
plt.figure(figsize=(16,6))

plt.subplot(2,2,1)
plt.scatter(study_hrs,marks,color='red')

plt.subplot(2,2,2)
plt.plot(study_hrs,marks,'b--')

plt.subplot(2,2,3)
plt.hist(marks,color='orange',rwidth=0.9)

plt.subplot(2,2,4)
plt.plot(study_hrs,marks,'g')
plt.plot(study_hrs,marks,'yo')
plt.savefig('D:\Analytics\subplot.png', facecolor='k') # To save the subplot
plt.show()
```
