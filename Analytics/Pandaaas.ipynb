### Filling the null values

```python
df.isnull().sum()
```

#### 1st method

```python
df1=df.fillna(0)  #Fills null with 0     # To save dff.fillna(0,inplace=True) 
df1.isnull().sum()
```

```python
df.fillna({'Physics':'0','Chemistry':'na'})    #To fill different values to null values of different columns
```

```python
df['Physics'].fillna(value=df['Physics'].mean())  # It fills the null values of "Physics" column with average of "Physics" column
```

#### 2nd method Forward fill

```python
df.fillna(method='ffill', axis=1) # It fills a null value with its previous value 
                                   # For axis=0 previous val is above val, For axis=1 previous val is left val.
```

#### 3rd method Backward fill

```python
df.fillna(method='bfill', axis=1) # It fills a null value with its next value 
df.fillna(method='bfill') # For axis=0 next val is below val, For axis=1 next val is right val.
```

## Merge() Function

```python
df1= pd.DataFrame({'Roll No':[1,2,3,4],'Physics':[20,20,30,40]})
df1
```

```python
df2= pd.DataFrame({'Roll No':[1,2,3,4],'Chemistry':[45,34,37,40]})
df2
```

```python
pd.merge(df1,df2,on='Roll No')
```

```python
pd.merge(df2,df1)  # by default merging columns is taken
```

```python
df3= pd.DataFrame({'Roll No':[1,2,3,4],'Physics':[20,20,30,40]})
df4= pd.DataFrame({'Roll No':[1,2,8,9],'Chemistry':[45,34,37,40]})
```

```python
pd.merge(df3,df4) # it takes rows of only intersecting columns
```

```python
pd.merge(df3,df4,how='left')
```

```python
pd.merge(df3,df4,how='right')
```

```python
pd.merge(df3,df4,how='outer')
```

## concat() function

```python
df1=pd.DataFrame({'Rollno':[1,2,3,4],'Maths':[30,40,50,60],'Physics':[33,44,55,66]})
df2=pd.DataFrame({'Rollno':[6,7,8,9],'Maths':[34,45,56,67],'Physics':[39,48,57,66]})
print(df1)
print(df2)
```

```python
pd.concat([df1,df2],ignore_index=True,sort=True)
# print(x)
```

```python
df1=pd.DataFrame({'Rollno':[1,2,3,4],'Maths':[30,40,50,60],'Chemistry':[33,44,55,66]})
df2=pd.DataFrame({'Rollno':[6,7,8,9],'Maths':[34,45,56,67],'Physics':[39,48,57,66]})
pd.concat([df1,df2],ignore_index=True)
```

```python
df1=pd.DataFrame({'Rollno':[1,2,3],'Maths':[30,40,50],'Physics':[33,44,55],'Chemistry':[55,66,77]})
df2=pd.DataFrame({'Rollno':[6,7,8],'Maths':[34,45,56],'Physics':[39,48,57]})
pd.concat([df1,df2],ignore_index=True)
```
