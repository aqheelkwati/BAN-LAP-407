```python
# import numpy as np
# import matplotlib.pyplot as plt # for visualisation
# %matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape') #encoding='unicode_escape' It is used to avoid errors during
                                                                  #Loading(Encoding error) of data, Its is optional
# df=pd.read_csv(r'D:\daily problems\project1\Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape   #Display the total number  of rows & columns
```

```python
df.head()  # Shows all data of 1st 5 rows
```

```python
df.info()
```

```python
# To remove blank columns
df.drop(['Status','unnamed1'],axis=1,inplace=True) # axis=1 means implement it on the entrire column with all rows
                                                   # inplace=True means save all the changes there itself
```

```python
# pd.isnull(df) #Shows which rows are null & which are not(it is not that usefull)
```

```python
# This Tells us that which column has how many null rows
pd.isnull(df).sum()
```

```python
#To remove null rows
df.dropna(inplace=True)
```

```python
#To change a datatype of any column
df['Amount']=df['Amount'].astype('int')
```

```python
#To check datatype of any column
df['Amount'].dtypes
```

```python
df.columns   #Displays list of all columns
```

# Exploratory data analysis

### Gender

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(6,5)})
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(6,5)})
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
x=sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

*From above graphs we can say that most of the buyers are females & even purchasing power of women are greater than men*

# Age

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_age=df.groupby(['Age Group','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Age Group',y='Amount',data=sales_age,hue='Gender')
# for bar in ax.containers:
#     ax.bar_label(bar)
```

*From above graphs we can say that most of the buyers are of age group 26-35 females*

### State

```python
#Total numbers of orders for top 10 states
# ax=sns.countplot(x='State',data=df)
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State',y='Orders',data=sales_state,)
```

```python
#Total numbers of sales amount for top 10 states
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

*From above graphs we can say that most of the Orders and sales/Amount are from States Uttar Pradesh,Maharashtra Then karnataka*

### Marital Status

```python

ax=sns.countplot(x='Marital_Status',data=df,hue='Gender')
sns.set(rc={'figure.figsize':(4,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_state,hue='Gender')
```

*From above Graphs we can say that Most of the buyers are Unmarried(Women) And they have high purchasing power*

### Occupation

```python
# df.columns
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Occupation',data=df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=sales_state)
```

*From above Graphs we can say that Most of the buyers are from IT,Healthcare & Aviation Sector*

### Product Category

```python
sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(x='Product_Category',data=df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python

sns.set(rc={'figure.figsize':(25,5)})
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=sales_state)
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

# Conclusion

*Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category*
